```python
import pandas as pd
import numpy as np
```

```python
diabetes_dataset = pd.read_csv('../data/diabetes.csv')
```

Data Collection and Analysis

PIMA Diabetes Dataset

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

Outcome 0 Means non-diabetic

Outcome 1 Means diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# separating the data and labels
# we use axis = 1 to say that we will drop a column not a row , for row we use 0
X = diabetes_dataset.drop(columns="Outcome", axis=1)
Y = diabetes_dataset['Outcome']
```

Data Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

Standardization (in data preprocessing) means scaling the features so that they have:

Mean = 0

Standard deviation = 1

```python
standarized_data = scaler.transform(X)
```

```python
X = standarized_data #data
Y = diabetes_dataset['Outcome'] #label , or desired Output
```

```python
print(X)
print(Y)
```

Split data into test and train

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

Training the Model

```python
from sklearn import svm
```

```python
classifier  =svm.SVC(kernel='linear')
```

```python
classifier
```

```python
# training the SVM classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
from sklearn.metrics import accuracy_score
```

```python
#Accuracy score on training data
x_training_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(x_training_prediction, Y_train)
```

```python
print(f'Accuracy score of the training data {training_data_accuracy}')
```

```python
#Accuracy score on training data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(f'Accuracy score of the test data {test_data_accuracy}')
```
